```python
def calculate_iou(box1, box2):
    # Lấy tọa độ của bounding box
    x1, y1, w1, h1 = box1
    x2, y2, w2, h2 = box2

    # Tính toán các điểm góc của bounding box
    x_left = max(x1 - w1 / 2, x2 - w2 / 2)
    y_top = max(y1 - h1 / 2, y2 - h2 / 2)
    x_right = min(x1 + w1 / 2, x2 + w2 / 2)
    y_bottom = min(y1 + h1 / 2, y2 + h2 / 2)

    # Tính toán diện tích của phần chồng lấn
    intersection_area = max(x_right - x_left, 0) * max(y_bottom - y_top, 0)

    # Tính toán diện tích của cả hai bounding box
    box1_area = w1 * h1
    box2_area = w2 * h2

    # Tính chỉ số IoU
    iou = intersection_area / (box1_area + box2_area - intersection_area)

    return iou
```

```python
box1 = [float(item) for item in input("Ground truth : ").split()]
```

```python
box2 = [float(item) for item in input("Bounding box: ").split()]
```

```python
print(calculate_iou(box1,box2))
```
